*Import the libraries*

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create a $10^4x10^4$ matrix, **A**, and a $10^4x1$ vector, **x** with random numbers.

```python
#Step 1 and 2
A = np.random.rand(10000, 10000)  # Create a random 10^4 x 10^4 matrix
x = np.random.rand(10000, 1)      # Create a random 10^4 x 1 vector
```

Perform a matrix-vector multiplication

```python
#Step 3
b = np.dot(A, x)  #matrix-vector multiplication
```

Make a scatter plot of the values of the vector b using matplotlib

```python
#Step 4
plt.figure(figsize=(14, 6))
plt.scatter(range(len(b)), b, s=2, c='navy')
plt.title('Scatter Plot of the values of Vector b')
plt.xlabel('Index')
plt.ylabel('Value of b')
#plt.xlim(left=0)
#plt.xlim(right=10000)
plt.show()
```
